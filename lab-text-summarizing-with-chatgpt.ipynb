# Text summarizing with ChaptGPT
In this lesson, you will summarize text with a focus on specific topics.

## Setup

````python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

## 1. Overview

In this lab, I explored how different prompting techniques can be used to summarize news-style texts related to mental health, sports, and personal development. The objective was to evaluate how the model behaves under different instructions, such as limiting word count, focusing on specific aspects, extracting key points, and summarizing multiple texts.

The experiment showed that the quality and relevance of the output depend heavily on how the prompt is structured, rather than only on the content of the text.

---

## 2. Prompt Variations and Results

### 🔹 Basic Summarization (Word Limit)

When asked to summarize the text in a limited number of words, the model was able to generate concise responses. However, in some cases, important details were omitted in order to meet the word constraint. This showed that strict limits can reduce the quality of the summary.

---

### 🔹 Focused Summarization

When the prompt specified a focus (e.g., mental health or social skills), the model successfully filtered the information and produced more relevant summaries. This demonstrated that adding constraints improves the usefulness of the output.


---

### 🔹 Multiple Text Summarization

When summarizing multiple articles at once, the model generated a more general and abstract response. While it captured the common ideas, it sometimes ignored specific details from individual texts. This showed that increasing the input size leads to more generalized outputs.

---

## 3. Variations That Did Not Work Well

Some prompt variations produced weaker results:

- When the instructions were too vague, the model generated generic summaries  
- When strict word limits were applied, key information was sometimes lost  
- When summarizing multiple texts, some articles were partially ignored  
- In some cases, the model slightly rephrased ideas instead of truly summarizing  

Although the model did not produce major factual errors, it occasionally oversimplified the content, which could be considered a form of information loss.

---

## 4. Key Learnings

From this lab, I learned that:

- The model responds more to structure than to intent  
- Clear and specific prompts significantly improve output quality  
- Extraction tasks are often more reliable than summarization  
- Summarizing multiple texts increases abstraction but reduces detail  
- Constraints (like word limits or focus) must be balanced to avoid losing key information
